```python
import numpy as np 
import pandas as pd

from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df =sns.load_dataset('penguins')
df.head(2)
```

```python
df = df.dropna(how='any')
df.info()
```

```python
bodymass = df['body_mass_g'].values
print(bodymass)
```

```python
sns.set(rc={'figure.figsize': (12,3)})
bm = sns.boxplot(x=bodymass)
```

```python
bm = df['body_mass_g'].values
mean = df['body_mass_g'].mean()
sns.histplot(x = bm)
bm = plt.axvline(mean,0, 1, color = 'red')
```

```python
bm = sns.lineplot(x = 'flipper_length_mm', y = 'body_mass_g', data = df, ci = None)
```

```python
bm = sns.lineplot(x = 'flipper_length_mm', y = 'body_mass_g', data = df, hue = 'sex', palette = ["blue","pink"], ci = None)
```

```python
bm = sns.boxplot(x = 'sex', y = 'body_mass_g', palette = ["blue","pink"], data = df)
sns.set(rc={'figure.figsize': (12, 3)})
```

```python
bm = sns.lineplot(x = 'flipper_length_mm', y = 'body_mass_g', data = df, hue = 'species', palette = ["purple", "green", "red"], ci = None)
```

```python
bm = sns.boxplot(x = 'species', y = 'body_mass_g', palette = ["purple", "green", "red"], data = df)
sns.set(rc={'figure.figsize': (12, 3)})
```

```python
bm = sns.lineplot(x = 'flipper_length_mm', y = 'body_mass_g', data = df, hue = 'island', palette = ["yellow", "teal", "crimson"], ci = None)
```

```python
bm = sns.boxplot(x = 'island', y = 'body_mass_g', palette = ["yellow", "teal", "crimson"], data = df)
sns.set(rc={'figure.figsize': (12, 3)})
```

```python
bm = sns.scatterplot(x = 'flipper_length_mm' , y = 'body_mass_g', data = df, color = 'b', marker = "o")
```
